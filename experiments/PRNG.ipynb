```python
state = 1111

def von_neumann():
    global state
    
    before = state
    
    square = str(state * state)
    
    h = len(square) // 2
    
    start = h - 1
        
    state = int(square[start:start + 4])
    
    if state == 0:
        state = before + 1

    return state / 9999
```

```python
x1 = 34
x2 = 45
def fibo():
    global x1, x2
    
    new_x = (x1 + x2) % (2 ** 16)
    x2 = x1
    x1 = new_x
    
    return new_x / (2 ** 16)
```

```python
MAX = 2 ** 31
```

```python
state = 1

def randu():
    global state
    state = 65539 * state % (2 ** 31)
    return state / MAX
```

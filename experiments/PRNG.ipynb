```python
state = 1111

def von_neumann():
    global state
    
    before = state
    
    square = str(state * state)
    
    h = len(square) // 2
    
    start = h - 1
        
    state = int(square[start:start + 4])
    
    if state == 0:
        state = before + 1

    return state / 9999
```

```python
MAX = 2 ** 31
```

```python
cyc = 1111

def von_neumann_cyc():
    global cyc
    
    cyc = cyc * cyc % MAX
    
    return cyc
```

```python
bad_state = 0x333

def bad():
    global bad_state
    bad_state = ((bad_state * 18694) + 1) % (2 ** 31)
    return bad_state / MAX
```

```python
x1 = 1
x2 = 1
def fibo():
    global x1, x2
    
    new_x = (x1 + x2) % (2 ** 31)
    x2 = x1
    x1 = new_x
    
    return new_x / (2 ** 31)
```

```python
state = 1

def randu():
    global state
    state = 65539 * state % (2 ** 31)
    return state / MAX
```

```python
state = 1

def randu_bin():
    global state
    state = 65539 * state % (2 ** 31)
    return int(bin(state)[2:10], 2)
```

```python
bs = 0

def bad_sedgewick():
    global bs
    bs = (31415821 * bs + 1) % (10 ** 8)
    return bs
```
